# Different Chart Types

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plot

### When to use Scatter Plot

- Scatter plots are ideal for showing the relationship between two variables. 
- They can help in identifying whether two variables are positively correlated, negatively correlated, or not correlated at all.

```python
np.random.seed(100000) # to keep the random value constant
x = np.random.randint(18, 66, 100)
y = 500 - (x * 3) + np.random.normal(0, 60, 100)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5)

plt.title('Age vs. Monthly Expenditure on Entertainment')
plt.xlabel('Age')
plt.ylabel('Monthly Expenditure ($)')

plt.show()
```

## Histogram

### When to use Histogram

- To see if the data follows a normal distribution.
- To quickly identify outliers that fall far from the rest of the distribution.

```python

ages = np.random.normal(loc=40, scale=10, size=1000) # Mean age 40, std deviation 10


plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, color='skyblue', edgecolor='r')

plt.title('Distribution of Ages in a Sample Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## Box Plot

```python
class1_scores = np.random.normal(loc=70, scale=10, size=100)
class2_scores = np.random.normal(loc=75, scale=8, size=100)
class3_scores = np.random.normal(loc=65, scale=15, size=100)

plt.figure(figsize=(10, 6))
plt.boxplot([class1_scores, class2_scores, class3_scores], patch_artist=True, labels=['Class 1', 'Class 2', 'Class 3'])
plt.title('Test Score Distribution Across Three Classes')
plt.ylabel('Scores')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 20, 22, 25, 28, 30, 35, 40, 45, 50, 55, 60, 70, 80, 150]

plt.boxplot(data)

plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot Example')
```

### Box Plot Elements

- Box: Represents the interquartile range (IQR) and contains the middle 50% of the data.
- Whiskers: Lines extending from the box that show the range of the data.
- Median Line: A line inside the box that represents the median (50th percentile) of the data.
- Outliers: Individual points beyond the whiskers are considered potential outliers and are often plotted individually.
- Minimum – It is the minimum value in the dataset excluding the outliers.
- First Quartile (Q1) – 25% of the data lies below the First (lower) Quartile.
- Third Quartile (Q3) – 75% of the data lies below the Third (Upper) Quartile.
- Maximum – It is the maximum value in the dataset excluding the outliers.

## Kernel Density Plot

- Seaborn
- Plotly

### When to use KDE Plot

- When you want to see the shape of the data distribution clearly, especially if the data is multimodal (has multiple peaks).
- KDE can be more visually informative than histograms for comparing the underlying distributions of different datasets, as it provides a smooth curve without the binning bias.

```python
import seaborn as sns

data = [15, 20, 22, 25, 28, 30, 35, 40, 45, 50, 55, 60, 70, 80, 150, 67]
sns.kdeplot(data)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Kernel Density Estimation (KDE) Plot')
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

ax = plt.figure().add_subplot(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)

# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5, rstride=8, cstride=8,
                alpha=0.3)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph
ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap='coolwarm')
ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap='coolwarm')
ax.contourf(X, Y, Z, zdir='y', offset=40, cmap='coolwarm')

ax.set(xlim=(-40, 40), ylim=(-40, 40), zlim=(-100, 100),
       xlabel='X', ylabel='Y', zlabel='Z')

plt.show()
```
